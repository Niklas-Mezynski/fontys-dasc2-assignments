## Exercise Countries of the World
The assignment is to find the 3 pairs of variables/columns that have the most strong relation. For example, you could, e.g. start with Gross Domestic Product (GDP), which is an indicator for a country's wealth and Infant mortality. Use pandas to read the data set, NumPy for the necessary data transformations and Matplotlib to inspect (just by visualisation, not computation) results.

```python
# Compute the correlation matrix
corr = df.corr()

corr
```

```python
# Plot a visual correlation matrix

import seaborn as sns
sns.set(style="white")

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
# cmap = 'PiYG'
cmap = sns.diverging_palette(280, -10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.9, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

```

```python
corr

# For the corr matrix, find the 3 highest (absolute) correlations and only show those

corr2=corr.abs()
highest_corr = corr2.unstack().sort_values(ascending=False).drop_duplicates()[1:5]

highest_corr
```

## Answer
To conclude the 3 variable pairs with the strongest correlation are (in order of strength):
- Other percentage and Arable percentage (-0.86)
- Birthrate and Infant mortality (0.84)
- Phones per 1000 people and GDP per capita (0.83)
